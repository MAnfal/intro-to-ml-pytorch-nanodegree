```python
'''
Let's plot this function.

f(x) = sin(x)x**2
'''
import matplotlib.pyplot as plt
import numpy as np

x_axis = np.arange(-10, 10, 0.01)

y_axis = [ np.sin(x) * x**2 for x in x_axis ]

plt.plot(x_axis, y_axis)

plt.show()
```

```python
'''
Let's plot this function.

f(x) = e**x
'''
import matplotlib.pyplot as plt
import numpy as np

x_axis = np.arange(-10, 10, 0.01)

y_axis = [ np.exp(x) for x in x_axis ]

plt.plot(x_axis, y_axis)

plt.show()
```

```python
'''
Let's plot this function.

f(x) = sin(x)
'''
import matplotlib.pyplot as plt
import numpy as np

x_axis = np.arange(-10, 10, 0.01)

y_axis = [ np.sin(x) for x in x_axis ]

plt.plot(x_axis, y_axis)

plt.show()
```

```python
'''
Let's plot this function.

f(x) = 2x^2 - x^3
'''
import matplotlib.pyplot as plt
import numpy as np

x_axis = np.arange(-10, 10, 0.01)

y_axis = [ 2 * x**2 - x**3 for x in x_axis ]

plt.plot(x_axis, y_axis)

plt.show()
```
