```python
import matplotlib.pyplot as plt
import numpy as np

'''
The area of the circle becomes more closer to the original value as dr_approaches 1.

This is proved below.
'''
dr_fine = 0.5
dr_coarse = 1

total_number_of_rings = 20

# Indicates the number of concentric rings I want to divide my circle into.
number_of_coarse_rings = np.arange(1, total_number_of_rings + 1, dr_coarse)
number_of_fine_rings = np.arange(1, total_number_of_rings + 1, dr_fine)

'''
If we slice a circle into concentric rings and expand the rings into thin strips, the strip will be like a rectangle.
The area of rectangle is w * h so we have the area of that ring = 2pir * dr where dr represents the width of the slice.

In this method we are only calculating the width of each ring when expanded and graphed in x and y axis.
'''
def get_y(number_of_rings):
    return [ 2 * np.pi * ring_number for ring_number in number_of_rings]

fig, (plt1, plt2) = plt.subplots(1, 2)

# Plot with fewer rings.
y_axis_coarse = get_y(number_of_coarse_rings)

plt1.set_title('Plot with fewer rings')
plt1.bar(number_of_coarse_rings, y_axis_coarse)
plt1.plot(number_of_coarse_rings, y_axis_coarse, color='red')


# Plot with more rings.
y_axis_fine = get_y(number_of_fine_rings)

plt2.set_title('Plot with more rings')
plt2.bar(number_of_fine_rings, y_axis_fine)
plt2.plot(number_of_fine_rings, y_axis_fine, color='red')


plt.show()

'''
Conclusion from this plotting experiment is as follows.

1. If we divide a circle in concentring rings.
2. Then slice those concentric rings to form long strips.
3. Then we can take the heights of those long strips as 2 * pi * r and the width of each ring as dr.
4. Increase in dr will give us coarser/fewer rings.
5. Decrease in dr will give us finer rings meaning we will be able to better anticipate the area of the circle.

Proof is given below.
'''

'''
Area of the circle is pi * r**2
'''
print(f"Area of the circle: { np.pi * ( total_number_of_rings ** 2 ) }")

'''
Formula to find area of a right triangle is 1/2 * base * height

1. base is the max of the number of rings.
2. height is the width of the biggest ring, coarse is chosen here to have a 1 - 1 mapping with the number of x-axis items we have.
'''
print(f"Area of the circle using coarse slicing right triangle: { 0.5 * len(number_of_coarse_rings) * np.max(y_axis_coarse) }")

'''
The base is divided by 2 because the density is twice now. We know density = m/v or mass/volume. Mass is 40 rings / available volume of 20.

The slight difference in the reading by fine is because the max is calculated at 20.5, if we can bring it closer to 20 then we will have a better result.
'''
print(f"Area of the circle using fine slicing right triangle: { 0.5 * ( len(number_of_fine_rings) / 2 ) * np.max(y_axis_fine) }")
```

```python
# Task: Plot a parabola.
import numpy as np
import matplotlib.pyplot as plt

x_axis = np.arange(-10, 10, 0.001)

def get_y(x_axis):
    return [x_coord**2 for x_coord in x_axis]

y_axis = get_y(x_axis)

plt.plot(x_axis, y_axis)

plt.show()
```
